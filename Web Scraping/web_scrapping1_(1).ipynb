```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from time import sleep
```

```python
base_url = 'https://www.1mg.com'

a = []
for i in range(1,2):

    url = f'{base_url}/categories/homeopathy-57?filter=true&pageNumber=={i}'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'}

    sleep(0.05)

    req = requests.get(url, headers = headers)
    soup = BeautifulSoup(req.content, 'html.parser')

    soup = BeautifulSoup(req.content)

    products = soup.find_all('div',class_='style__product-box___liepi')



    for product in products:


        url = product.find('a',class_='style__product-link___UB_67').get('href')
        link = 'https://www.1mg.com'+url
        a.append(link)
    sleep(0.05)
```

```python
import time
from time import sleep
```

```python
time_limit = 1
Name = []
Brand_name = []
Key_benefits = []
Key_ingredients = []
Rating = []
Number_of_rating = []

for link in a:
    req1 = requests.get(link, headers=headers)
    soup1 = BeautifulSoup(req1.text, 'html.parser')

    name = soup1.find('h1',class_='ProductTitle__product-title___3QMYH')
    name=name.text.strip() if name else ""
    Name.append(name)

    Brand1 = soup1.find('div',class_='ProductTitle__manufacturer___sTfon')
    Brand = Brand1.text.strip() if Brand1 else ""
    Brand_name.append(Brand)

    benefit = soup1.find('strong',string='Key Benefits:')
    benefits = benefit.find_next('li').string.strip() if benefit else ''
    Key_benefits.append(benefits)

    bag = soup1.find('strong', string = 'Key Ingredients:')
    key_ingredients = bag.find_next('li').string.strip() if bag else ''
    Key_ingredients.append(key_ingredients)

    ratings = soup1.find('div',class_='RatingDisplay__ratings-container___3oUuo')
    ratings=ratings.text.strip() if ratings else ""
    Rating.append(ratings)

    rating_number = soup1.find('span', class_='RatingDisplay__ratings-header___ZNj5b')
    rating_number=rating_number.text.strip() if rating_number else ""
    Number_of_rating.append(rating_number)
    time.sleep(0.05)



data ={'Name':Name,'Brand_name':Brand_name,'Key_benefits':Key_benefits,'Key_ingredients':Key_ingredients, 'Rating':Rating,'Number_of_rating':Number_of_rating}

df = pd.DataFrame(data)
df
```
